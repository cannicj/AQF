```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import datetime

path = 'DATA/' 
```

```python
VRP_df = pd.read_csv(path+'VRP_data.csv',usecols = ['date','RV^2','VIX^2'])
VRP_df['date'] = pd.to_datetime(VRP_df['date'],format = '%Y-%m-%d')
```

```python
VRP_df
```

```python
VRP_df['RV^2_next'] = VRP_df['VIX^2'].shift(-21)
VRP_df = VRP_df.dropna()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = VRP_df.loc[:,['RV^2','VIX^2']]
Y = VRP_df.loc[:,'RV^2_next']
    
reg = LinearRegression().fit(X, Y)

Y_fitted = reg.predict(X)
VRP_df['E_RV^2'] = Y_fitted

VRP_df['VRP'] = VRP_df['VIX^2'] - VRP_df['E_RV^2']
```

```python
VRP_df
```

```python
VRP_df[['date','VRP']].to_csv(path+'VRP_data.csv')
```
